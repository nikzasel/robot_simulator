```python
import numpy as np
from skimage import draw, measure, data, color
import matplotlib.pyplot as plt
import random
```

```python
image = np.zeros((512, 512))
for i in range(random.randint(2, 5)):
    circle_center_y = random.randint(0, image.shape[0])
    circle_center_x = random.randint(0, image.shape[1])
    circle_radius = random.randint(32, 128)
    rr, cc = draw.circle(circle_center_y, circle_center_x, circle_radius, image.shape)
    image[rr, cc] = i
plt.imshow(image)
```

```python
image_color_selected = image>0
plt.imshow(image_color_selected)
```

```python
image_label = measure.label(image_color_selected)
plt.imshow(image_label)
```

```python
for region_prop in measure.regionprops(image_label):
    print(region_prop.area/image.size)
```
